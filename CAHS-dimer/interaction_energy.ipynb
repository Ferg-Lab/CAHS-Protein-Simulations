```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
```

```python
def read_xvg(file):
    data = np.loadtxt(file,comments=('@','#'))
    return data
```

```python
for i in range(5):
    data_LJ = read_xvg(f'./rerun/energy_lj{i+1}.xvg')
    data_Q = read_xvg(f'./rerun/energy_q{i+1}.xvg')

    total_ie = data_LJ[:,1] + data_Q[:,1]
    y_smooth = gaussian_filter1d(total_ie,sigma=1)
    plt.plot(data_Q[:,0]/1000, y_smooth/4.184, label=i+1)
plt.legend()
plt.xlabel('Time (ns)')
plt.xlabel('Interaction Energy (kcal/mol)')
```
